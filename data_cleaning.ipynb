Part 1: Data Cleaning

```python
# import libraries
import pandas as pd
```

```python
df = pd.read_excel("ramen_rating_2023.xlsx")
df
```

```python
#check null and clean null dataset
df.isnull().sum()
```

```python
df = df.dropna(subset=['Stars'])
df
```

```python
#drop columns not need
df = df.drop(['Review #', 'T'], axis=1)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.dtypes
```

```python
#clean column Star
df['Stars'].value_counts()
```

```python
#convert column Stars to numeric
df['Stars']=pd.to_numeric(df['Stars'], errors='coerce')
df['Stars'].value_counts()
```

```python
#remove all Nah
df = df.dropna()
df.isnull().sum()
```

```python
#test df
df
```

```python
#convert dataframe to csv
df.to_csv('ramen_rating_2023_clean.csv', index=None)
```
